```python
# Install openpyxl and xlrd if you haven't already
%pip install openpyxl xlrd

import pandas as pd

# Function to convert .xls to .xlsx
def convert_xls_to_xlsx(input_path, output_path):
    # Read the .xls file
    df = pd.read_excel(input_path, engine='xlrd')
    
    # Save it as .xlsx
    df.to_excel(output_path, index=False, engine='openpyxl')
    print(f"File converted and saved to {output_path}")

# Example usage
# file_path = "data/"
file_path = "backend/data/"

your_file = file_path + "UW_Churn_Pred_Data.xls"
your_new_file = file_path + "UW_Churn_Pred_Data.xlsx"


# Replace 'your_file.xls' and 'your_new_file.xlsx' with your actual filenames
convert_xls_to_xlsx(your_file, your_new_file)
```
